```python
train_idxs = features['is_listened'].notnull()
train = features[train_idxs].copy()
test = features[~train_idxs].copy()
```

```python
from collections import Counter

import matplotlib.pyplot as plt
import xam


feature = 'user_listen_count'

ewb = xam.preprocessing.EqualFrequencyBinner(n_bins=20)
ewb.fit(test[[feature]])
train['bin'] = ewb.transform(train[[feature]])
bin_counts = Counter(train['bin'])
train['weight'] = train['bin'].apply(lambda x: 1 / bin_counts[x])
sample = train.sample(800000, weights='weight')


fig, ax = plt.subplots(figsize=(12, 8))
train[feature].hist(bins=500, normed=True, ax=ax, alpha=0.3)
test[feature].hist(bins=500, normed=True, ax=ax, alpha=0.3)
sample[feature].hist(bins=500, normed=True, ax=ax, alpha=0.3)
ax.set_xlim([0, 2000]);
ax.set_ylim([0, 0.003]);
```
